# Mini Project on App Store Data Set

```python
import pandas as pd
filePath='AppleStore.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
readCsvData(filePath)
```

```python
df=readCsvData('AppleStore.csv')
```

### Problem Statement :-
- 1.To Calucate the average rating of Gaming and Non-Gaming Apps

### Problem Description :-
- In order to get the averages of the user ratings of the apps we need to use the df function with the help of the prime_genre column
- Then use the mean function to print their means

```python
tr=df[df['prime_genre'] == 'Games']
print("Average user rating of Gaming Apps is ",tr['user_rating'].mean())
ar=df[df['prime_genre'] != 'Games']
print("Average user rating of Non-Gaming Apps is ",ar['user_rating'].mean())
```

### Problem Statement :-
- 2.To draw a pie chart based on the following labels :-
 - Price=0 -> Free Apps
 - Price<20 -> Affordable Apps
 - Price<50 -> Expensive Apps
 - Price>50 -> Very Expensive Apps
### Problem Description :-
- In order to print a pie chart according to the prices of the apps we need to first obtain the number of apps in each range.
- For that first we need to use the df function with the required conditions.
- After obtaining the required elements we need to find thier count by using the len() function.
- Using the length values and the names of the categories we can draw the pie chart with the required condition.

```python
import matplotlib.pyplot as plt
a=len(df[df['price'] == 0])
b=len(df[(df['price'] >= 1) & (df['price'] <= 20)])
c=len(df[(df['price'] >= 21) & (df['price'] <= 50)])
d=len(df[(df['price'] >= 50)])
app=['Free','Very Expensive','Affordable','Expensive']
coun=[a,d,b,c]
plt.pie(coun,labels=app,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='lower left')
plt.show()
```

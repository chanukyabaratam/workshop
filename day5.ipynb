```python
li=[0,1,2,3,4,5]
for i in range(len(li)):
    print(li[i],end=' ')
print()
for i in li:
    print(i,end=' ')
```

```python
li=[1,2,3,4,5]
for i in range(len(li)):
    print(i,'---->',li[i])
print()
#pythonic way
for i,a in enumerate(li):
    print(i,'----->',a)
```

```python
li=[1,2,3,4,5]
for i in sorted(li):
    print(i,end=' ')
print()
for i in sorted(li,reverse=True):
    print(i,end=' ')
```

# Lambda functuins
 - anonymous function means that a function is without a name
 - any lambda function on python will be defined with lambda
 - syntax: lambda arg:expression

```python
def square(n):
    return n*n
square(10)
```

```python
a=lambda x :x*x
print(a(10))
```

### use of lambda with filter

```python
def filterlist(li):
    a=[]
    for i in li:
        if i%2==0:
            a.append(i)
    return a
li=[1,2,3,4,5,6] 
filterlist(li)
```

```python
li=[1,2,3,4,5,6] 
lam_li=list(filter(lambda x:(x%2==0),li))
print(lam_li)
```

```python
li=[1,2,3,4,5,6] 
map_list=list(map(lambda x:(x*2),li ))
print(map_list)
```

```python
def squarelist(li):
    a=[]
    for i in li:
        a.append(1*2)
    return a
li=[1,2,3,4,5,6] 
squarelist(li)
```

### lambda with reduce()

```python
from functools import reduce
li=[1,2,3,4,5,6] 
a=reduce((lambda x,y : x+y),li)
print(a)
```
